## Matplotlib example workflow 

```python
# 0. import matplotlib and get it ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1,2,3,4]
y = [11, 22, 33, 44]

# 2. Setup plot
fig, ax = plt.subplots(figsize = (15, 10)) # width and height of figure

# 3. Plot the data
ax.plot(x,y)

# 4. Customise plot
ax.set(title = "Simple plot",
      xlabel = "x-axis",
      ylabel = "y-axis")

# 5. Save & show (you save the whole figure)
fig.savefig('images/sample-plot.png')
```

### Two options for subplots

```python
# subplot option 1 (tuple way)
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows = 2,
                                             ncols = 2,
                                             figsize = (10, 5))

# Plot to each different axis
x = np.linspace(0,10,100)
ax1.plot(x, np.sin(x))
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.hist(np.random.randn(100))
ax4.bar(nut_butter_prices.keys(), nut_butter_prices.values())
```

```python
# Subplot option 2 (list index way)
fig, ax = plt.subplots(nrows = 2,
                       ncols = 2,
                       figsize = (10, 5))
# plot to each different index
ax[0, 0].plot(x, np.sin(x))
ax[0, 1].scatter(np.random.random(10), np.random.random(10))
ax[1, 0].hist(np.random.randn(100))
ax[1, 1].bar(nut_butter_prices.keys(), nut_butter_prices.values());
```

### Which one should you use? (pyplot vs matplotlib OO method?)

* When plotting something quickly, okay to use the pyplot method
* When plotting something more advanced, use the OO method

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
len(heart_disease), len(over_50)
```

```python
# pyplot method aka plotting directly from panda's df
over_50.plot(kind = 'scatter',
             x = 'age', 
             y = 'chol',
             c = 'target',
             figsize = (6, 3));
```

```python
## OO method mixed with pyplot method
fix, ax = plt.subplots(figsize = (6, 3))
over_50.plot(kind = 'scatter',
             x = 'age',
             y = 'chol',
             c = 'target',
             ax = ax);

ax.set_xlim([45, 100]);
```

```python
## OO method from scratch
fix, ax = plt.subplots(figsize = (10, 6))

# Plot the data

scatter = ax.scatter(x = over_50["age"],
                     y = over_50["chol"],
                     c = over_50["target"]);

# Customise
ax.set(title = "Heart Disease and Choresterol Levels",
       xlabel = "Age",
       ylabel = "Choresterol");

# Add a legend
ax.legend(*scatter.legend_elements(), title = "Target");  # '*' tells to grab the values from "scatter" and unpack legend
                                                          # elements (from c = over_50["target"]) 
                                                          # and create elements built off these elements
        
# Add a horizontal line
ax.axhline(over_50["chol"].mean(),
           linestyle = '--');
```

```python
over_50.head()
```

```python
# subplot of chol, age, thalach

fig, ax = plt.subplots(nrows = 2, ncols = 1,
                       figsize = (10, 6),
                      sharex= True)

# Add data to ax[0]
scatter =  ax[0].scatter(x = over_50["age"],
                          y = over_50["chol"],
                          c = over_50["target"])
# Customise ax[0]
ax[0].set(title = "Heart Disease and Choresterol Levels",
          ylabel = "Choresterol");
# Add a legend to ax[0]
ax[0].legend(*scatter.legend_elements(), title = "Target");
# Add a horizontal line to ax[0]
ax[0].axhline(over_50["chol"].mean(),
             linestyle = '--');

# Add data to ax[1]
scatter1 = ax[1].scatter(x = over_50["age"],
                         y = over_50["thalach"],
                         c = over_50["target"])
# customise ax[1]
ax[1].set(title = "Heart Disease and Maximum Heart Rate",
         xlabel = "Age",
         ylabel = "Max Heart Rate");
# Add a legend to ax[1]
ax[1].legend(*scatter.legend_elements(), title = "Target");
# Add a horizontal line to ax[1]
ax[1].axhline(over_50["thalach"].mean(),
             linestyle = '--');

# Add the title to the figure
fig.suptitle("Heart Disease Analysis", fontsize = 16, fontweight = "bold");
```
